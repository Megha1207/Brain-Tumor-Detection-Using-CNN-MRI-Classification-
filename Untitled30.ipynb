```python
train_dir = "/content/mri_split/train"
val_dir   = "/content/mri_split/val"

img_size = (150, 150)
batch_size = 16

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

val_data = val_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
import os, shutil

src = "/content/mri/brain_tumor_dataset"
dst = "/content/mri/brain_tumor_dataset_binary"

os.makedirs(os.path.join(dst,"tumor"), exist_ok=True)
os.makedirs(os.path.join(dst,"no"), exist_ok=True)

# Move 'no' images
for f in os.listdir(os.path.join(src,"no")):
    shutil.copy(os.path.join(src,"no",f), os.path.join(dst,"no"))

# Merge 'yes' + 'Y' as tumor
for folder in ["yes", "Y"]:
    folder_path = os.path.join(src, folder)
    if os.path.exists(folder_path):
        for f in os.listdir(folder_path):
            shutil.copy(os.path.join(folder_path,f), os.path.join(dst,"tumor"))
```

```python
import splitfolders
splitfolders.ratio(
    dst,
    output="/content/mri_split",
    seed=42,
    ratio=(0.8,0.2)
)
```

```python
train_dir = "/content/mri_split/train"
val_dir   = "/content/mri_split/val"

train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)

val_data = val_datagen.flow_from_directory(
    val_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='binary'
)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(150,150,3)),
    tf.keras.layers.MaxPool2D(2,2),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),

    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPool2D(2,2),

    tf.keras.layers.Flatten(),

    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dropout(0.3),

    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
history = model.fit(
    train_data,
    validation_data=val_data,
    epochs=10
)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model

def grad_cam_auto(model, img_path, last_conv_layer_name=None, img_size=(150,150)):
    """
    Grad-CAM for Sequential model. Automatically initializes layers.

    Args:
        model: Trained Keras Sequential model
        img_path: Path to input MRI image
        last_conv_layer_name: Name of last conv layer (optional). If None, auto-detects.
        img_size: Target size of input image
    """
    # Load and preprocess image
    img = image.load_img(img_path, target_size=img_size)
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0) / 255.0

    # Automatically pick last conv layer if not provided
    if last_conv_layer_name is None:
        for layer in reversed(model.layers):
            if isinstance(layer, tf.keras.layers.Conv2D):
                last_conv_layer_name = layer.name
                break
        if last_conv_layer_name is None:
            raise ValueError("No Conv2D layer found in model for Grad-CAM.")

    # Get the last conv layer
    last_conv_layer = model.get_layer(last_conv_layer_name)

    # Create a functional model from input to last conv layer
    # This bypasses the Sequential model's output issue
    last_conv_layer_model = Model(model.inputs, last_conv_layer.output)

    # Create a model for the classifier part (from last conv to output)
    classifier_input = tf.keras.Input(shape=last_conv_layer.output.shape[1:])
    x = classifier_input

    # Find the index of the last conv layer
    last_conv_idx = None
    for idx, layer in enumerate(model.layers):
        if layer.name == last_conv_layer_name:
            last_conv_idx = idx
            break

    # Apply all layers after the last conv layer
    for layer in model.layers[last_conv_idx + 1:]:
        x = layer(x)

    classifier_model = Model(classifier_input, x)

    # Compute Grad-CAM
    with tf.GradientTape() as tape:
        conv_outputs = last_conv_layer_model(img_array)
        tape.watch(conv_outputs)
        predictions = classifier_model(conv_outputs)
        class_idx = np.argmax(predictions[0])
        loss = predictions[:, class_idx]

    # Compute gradients of the target class w.r.t conv output
    grads = tape.gradient(loss, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0,1,2))

    # Weight conv outputs with the gradients
    conv_outputs = conv_outputs[0]
    heatmap = conv_outputs @ pooled_grads[..., tf.newaxis]
    heatmap = tf.squeeze(heatmap)

    # Normalize
    heatmap = tf.maximum(heatmap, 0) / tf.math.reduce_max(heatmap)
    heatmap = heatmap.numpy()

    # Resize heatmap to match original image size
    heatmap = tf.image.resize(heatmap[..., np.newaxis], img_size).numpy()

    # Plot heatmap on original image
    plt.figure(figsize=(6,6))
    plt.imshow(img)
    plt.imshow(heatmap, cmap='jet', alpha=0.4)  # Overlay heatmap
    plt.title(f"Grad-CAM: Predicted class {class_idx}")
    plt.axis('off')
    plt.show()

    return heatmap, class_idx

# Example usage
sample_img_path = "/content/mri_split/val/yes/Y106.jpg"
grad_cam_auto(model, sample_img_path)
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense, Dropout, Concatenate
from tensorflow.keras.models import Model

# Define inputs with custom names
cnn_input = Input(shape=(150, 150, 3), name='cnn_input')
ehr_input = Input(shape=(3,), name='ehr_input')

# CNN branch
x = Conv2D(32, (3,3), activation='relu')(cnn_input)
x = MaxPooling2D(2,2)(x)
x = Conv2D(64, (3,3), activation='relu')(x)
x = MaxPooling2D(2,2)(x)
x = Conv2D(128, (3,3), activation='relu')(x)
x = MaxPooling2D(2,2)(x)
x = Flatten()(x)
x = Dropout(0.5)(x)
cnn_features = Dense(128, activation='relu')(x)

# EHR branch
ehr_features = Dense(32, activation='relu')(ehr_input)
ehr_features = Dense(16, activation='relu')(ehr_features)

# Fusion
combined = Concatenate()([cnn_features, ehr_features])
z = Dense(64, activation='relu')(combined)
z = Dropout(0.3)(z)
output = Dense(1, activation='sigmoid')(z)

# Create model
fusion_model = Model(inputs=[cnn_input, ehr_input], outputs=output)
fusion_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Now your original generator will work
def fusion_generator(img_gen, ehr_df):
    while True:
        imgs, labels = next(img_gen)
        idx = np.random.randint(0, len(ehr_df), size=len(labels))
        ehr_batch = ehr_df.iloc[idx].values.astype(np.float32)
        labels = labels.astype(np.float32)
        yield ({"cnn_input": imgs, "ehr_input": ehr_batch}, labels)

output_signature = (
    {
        "cnn_input": tf.TensorSpec(shape=(None, 150,150,3), dtype=tf.float32),
        "ehr_input": tf.TensorSpec(shape=(None, 3), dtype=tf.float32)
    },
    tf.TensorSpec(shape=(None,), dtype=tf.float32)
)

train_ds = tf.data.Dataset.from_generator(
    lambda: fusion_generator(train_data, ehr_features),
    output_signature=output_signature
)

val_ds = tf.data.Dataset.from_generator(
    lambda: fusion_generator(val_data, ehr_features),
    output_signature=output_signature
)

fusion_model.fit(
    train_ds,
    steps_per_epoch=len(train_data),
    validation_data=val_ds,
    validation_steps=len(val_data),
    epochs=10
)
```
